```python
import requests
from bs4 import BeautifulSoup
import redis

# Function to scrape programming languages from Wikipedia
def scrape_programming_languages():
    url = "https://en.wikipedia.org/wiki/List_of_programming_languages"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all language links in the list
    languages = []
    for li in soup.select('.div-col li'):
        language = li.text.strip()
        link = li.find('a')['href'] if li.find('a') else None
        languages.append({
            'language': language,
            'link': f"https://en.wikipedia.org{link}" if link else 'No link'
        })

    return languages

# Sample usage
languages = scrape_programming_languages()

for lang in languages[:5]:  # Display the first 5 languages
    print(lang)
```
